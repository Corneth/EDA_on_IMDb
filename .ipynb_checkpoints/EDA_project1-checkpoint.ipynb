```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from requests import get
from random import randint
from time import sleep
from bs4 import BeautifulSoup
```

```python
"""My stab at pulling information from the actual IMDb website."""
pages = np.arange(1,1001,250)
title = []
movie_rating = []
year =  []
star_rating = []
gross = []
movie_genres = []
movie_genre = []
movie_time = []
cast = []
director = []
def scraper():
    for page in pages:
        html = 'https://www.imdb.com/search/title/?release_date=1950-01-01,2020-12-31&groups=top_1000&sort=alpha,asc&count=250'#&start=' +str(page) +'&ref_=adv_nxt'
        response = get(html)
        soup = BeautifulSoup(response.text, 'html.parser')
        
#pulling from lister-item mode-advanced pulls all the information for each of the movies on the page to easily pull the tags as attributes of an item
        container = soup.findAll('div',class_ = 'lister-item mode-advanced') 
        """ for loop looking for each of the catagories in each of the containers"""
        for name in container:
            """nv is so we can seperate the gross income and seperate it from the votes set into their own """
            nv = name.find_all('span',attrs ={'name':'nv'})
            
            title.append(name.h3.a.text)
            movie_rating.append(name.p.span.text)
            year.append(name.h3.find('span',class_ = 'lister-item-year').text)
            
            star_rating.append(float(name.strong.text))
            
            gross_income = nv[1].text if len(nv) > 1 else '-'
            gross.append(gross_income)
            
            movie_genres.append(name.p.find('span',class_ ='genre').text)
            
            
            movie_time.append(name.p.find('span', class_ = 'runtime').text)
            #actors = name.p.find('a', class_ ='').text # throws error for NoneType
            #cast.append(actors)
            
            #director.append(cast[0])
scraper()
```
